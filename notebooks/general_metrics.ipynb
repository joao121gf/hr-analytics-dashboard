```python
import pandas as pd
from IPython.display import display
from datetime import datetime
df = pd.read_csv('../data/filtred/filtred_hr_dataset.csv')
display(df.head(2))
```

- Quantidade de funcion√°rios ativos e desligados

```python
# Conta ativos e desligados
total_ativos = df[df['Termd'] == 0].shape[0]
total_desligados = df[df['Termd'] == 1].shape[0]

# Cria o novo DataFrame com colunas 'M√©trica' e 'Valor'
general_df2 = pd.DataFrame({
    'M√©trica': ['Total de Funcion√°rios Ativos', 'Total de Funcion√°rios Desligados'],
    'Valor': [total_ativos, total_desligados]
})

display(general_df2)

inatives = df[df['Termd'] == 0]
```

```python
display(df[df['Termd'] == 1])
```

- Tempo m√©dio de perman√™ncia

```python
# Data de admiss√£o est√° na coluna: DateofHire
# Data de encerramento (desligamento) est√° na coluna: DateofTerminationDateofHire 
```

```python
print((df['DateofHire'].dtype))
print((df['DateofTermination'].dtype))
```

```python
df['DateofHire'] = pd.to_datetime(df['DateofHire'], errors='coerce')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'], errors='coerce')

```

```python
print((df['DateofHire'].dtype))
print((df['DateofTermination'].dtype))
```

```python
inatives = df[df['Termd'] == 1].copy()  
atives = df[df['Termd'] == 0].copy()

inatives.loc[:, 'DateofHire'] = pd.to_datetime(inatives['DateofHire'], errors='coerce')
inatives.loc[:, 'DateofTermination'] = pd.to_datetime(inatives['DateofTermination'], errors='coerce')

inatives.loc[:, 'time_permanency'] = (inatives['DateofTermination'] - inatives['DateofHire']).dt.days
display(inatives.head(2))
```

```python
general_df2.loc[len(general_df2)] = ['Tempo medio permanencia', int(inatives['time_permanency'].mean())]
general_df2['Valor'] = general_df2['Valor'].astype(int)
```

```python
display(general_df2)
```

- √çndice¬†de¬†Absente√≠smo

```python
# Apenas funcion√°rios ativos
ativos = df[df['Termd'] == 0].copy()

# Calcular tempo de perman√™ncia em dias at√© hoje
from datetime import datetime
today = pd.to_datetime('today')

ativos['DateofHire'] = pd.to_datetime(ativos['DateofHire'], errors='coerce')
ativos['days_in_company'] = (today - ativos['DateofHire']).dt.days

# Calcular faltas por m√™s (normalizado)
ativos['absences_total_days'] = ativos['Absences']  # total acumulado
ativos['absences_per_month'] = ativos['absences_total_days'] / (ativos['days_in_company'] / 30)

# Converter faltas por m√™s em horas
ativos['absences_hours_per_month'] = ativos['absences_per_month'] * 8

# Total de horas previstas por m√™s por funcion√°rio
hora_prev_mensal = 176

# √çndice m√©dio de absente√≠smo (%)
ativos['abs_percent'] = (ativos['absences_hours_per_month'] / hora_prev_mensal) * 100
mean_abs = round(ativos['abs_percent'].mean(), 2)


display(mean_abs)
```

```python
# general_df2.loc[len(general_df2)] = ['Tempo medio permanencia', int(inatives['time_permanency'].mean())]
# general_df2['Valor'] = general_df2['Valor'].astype(int)


general_df2.loc[len(general_df2)] = ['√çndice de Absente√≠smo', mean_abs]
display(general_df2)
```

```python
# M√©dia de atrasos nos √∫ltimos 30 dias (`DaysLateLast30`)
display(df.head(1))
mean_days_late_last_30 = round(df['DaysLateLast30'].mean(), 2)
display(mean_days_late_last_30)
```

```python
general_df2.loc[len(general_df2)] = ['M√©dia de atrasos nos √∫ltimos 30 dias', mean_days_late_last_30]
display(general_df2)
```

```python
# Exportando df para csv
general_df2.to_csv('general_metrics.csv', index=False)
```
